Since in 1960 total population was low and maybe thats the reason for lower crime. So we need a different scale to find the maximum and minimum total crime year

```python
temp = df.loc[df.Total==min(df.Total)][['Population','Total','Year']]
temp['ratio'] = temp['Total']/temp['Population']
temp
```

```python
temp = df.loc[df.Total==max(df.Total)][['Population','Total','Year']]
temp['ratio'] = temp['Total']/temp['Population']
temp
```

No we were wrong. The total crime and population ration of 1991 is 3 times than the 1961. Which means actual crime rate has been increased.

```python
temp = df.loc[:,['Year','Total','Population']]

temp['crime_rate'] = temp['Total']/temp['Population']
temp
```

If you look closly to the crime rate and Year. From 1960 to 1980 is incresed constantly and then decreased till 1985. From 1980 to 1991 it start increasing again. After 1991 the crime rate is constantly decreases till 2014.
